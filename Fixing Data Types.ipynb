```python
new_types = {}
for col in list(teams_totals.columns):
    if col[-3:len(col)] == 'tot':
        new_types[col] = 'int64'
    elif col[0:4] == 'WINS':
        new_types[col] = 'int64'
    elif col[0:3] == 'LOS':
        new_types[col] = 'int64'
    else:
        continue
new_types['TEAM_ID'] = 'int64'

new_teams_totals = teams_totals.astype(new_types)
new_teams_totals.head()
```
